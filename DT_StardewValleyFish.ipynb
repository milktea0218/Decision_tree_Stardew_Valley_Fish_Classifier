```python
### **導入訓練資料集**
train_set = pd.read_excel("train_set.xlsx")
test_set = pd.read_excel("test_set.xlsx")
#print(tabulate(train_set, headers='keys', tablefmt='firstrow',showindex=False, numalign="left"))
```

#### **把訓練集視覺化一部分看看**

### **3.編碼**
####  **做one hot會產生太多欄位，所以我選擇做Label Encoding。 這邊直接導入處理完成的資料集即可**

```python
le_train_set = pd.read_excel("LE_train_set.xlsx")
le_test_set = pd.read_excel("LE_test_set.xlsx")
```

####  **地點:**
  >**海洋=1 ，山區=2, 河流=3**
####  **時間:**
  >**這邊我直接把小時分鐘轉成對應的數字**
####  **季節:**
  >**春=1 ，夏=2, 秋=3, 冬=3**
####  **天氣:**
  > **晴天=1 ，雨天=2**
####  **名稱:**
  >**河豚=1 ，鯷魚=2, 金槍魚=3, 虹鱒魚=4, 大嘴鱸魚=5, 太陽魚=6, 鯰魚=7, 鰻魚=8**

```python
le_train_set
```

#### **切分X跟Y**

```python
X_train = le_train_set.iloc[:, 0:-1]
y_train = le_train_set.iloc[:, -1]
X_test = le_test_set.iloc[:, 0:-1]
y_test = le_test_set.iloc[:, -1]
```

### **4.導入決策樹模型**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)
```

#### **印出精準度，看起來還可以**

```python
model.score(X_test, y_test)
```

#### **自己生一筆資料測試看看**
* #### **想知道處於 河流(3)、中午12:00、冬天(4)、雨天、體型22吋的魚是?  (答案應為鯰魚)**

```python
print(model.predict([[3,1200,4,2,22]])[0])
```

#### **得到 7 =鯰魚，答對~**
